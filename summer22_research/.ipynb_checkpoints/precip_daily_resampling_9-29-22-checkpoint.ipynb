### Downsampling 6 hourly precipitation accumulation data into daily data (1 day as 06z,12z,18z,0z)

```python
six_hourly=xr.open_dataset('/project/xprecip/merra2_ncfiles/precip/resampled/all_merra_precip_6hourly_7-12.nc')
six_hourly
```

```python
six_hourly_precip=six_hourly['PRECTOTCORR']
six_hourly_precip
```

### First, test the method with smaller data for one point

```python
#Take out first value

shifted=six_hourly_precip.loc['1980-01-01T06:00:00.000000000':'1980-01-03T00:00:00.000000000']
```

```python
winter_shifted=shifted.isel(time=shifted.time.dt.season.isin(['DJF']))
winter_shifted
```

```python
#dummy resample
zeros=winter_shifted-winter_shifted
dummy_resample=zeros.resample(time='1D').sum(dim='time')
dummy_resample
```

```python
resampled_times=dummy_resample['time']
resampled_times_accum=resampled_times[0:2]
```

```python
lon=winter_shifted['lon']
lat=winter_shifted['lat']
```

```python
og_precip_dc=winter_shifted.sel(lat=39,lon=-77,method='nearest')
og_precip_dc
```

```python
for i in range(0,len(og_precip_dc),4):
    print(np.nansum(og_precip_dc[i:i+4])) #prototype works!!!
```

```python
for i in range(0,len(og_precip_dc),4):
    print((og_precip_dc[i:i+4]).sum(dim='time')) #prototype works!!!
```

```python
# now building on this
resampled_list=[]
for i in range(0,len(winter_shifted),4):
    summation=(winter_shifted[i:i+4]).sum(dim='time')
    resampled_list.append(summation)
```

```python
resampled_list
```

```python
precip_array_test=xr.DataArray(resampled_list,dims=('time','lat','lon'),coords={'time':resampled_times_accum,'lon':lon,'lat':lat})
```

```python
precip_array_test
```

```python
dc=precip_array_test.sel(lat=39,lon=-77,method='nearest')
dc #it worked!!
```

### Now try opening the saved dataset

```python
f= '/project/xprecip/merra2_ncfiles/precip/resampled/merra_precip_daily_full.nc'
daily_prcp=xr.open_dataset(f)
lon=daily_prcp['lon']
lat=daily_prcp['lat']
daily_prcp
```

```python
daily_values=daily_prcp['__xarray_dataarray_variable__']
daily_values
```

```python
conus_mean=daily_values.mean(dim='time')
conus_mean
```

```python
fig = plt.figure(figsize=(11,8.5))
ax=plt.axes(projection=ccrs.Robinson())
cs=ax.contourf(lon, lat, conus_mean,transform = ccrs.PlateCarree(),cmap='YlGnBu',extend='both')
ax.set_extent([-124.5,-66.8,23.486,49])
ax.coastlines()
ax.gridlines()
cbar=plt.colorbar(cs, shrink=0.75)
plt.title('Daily Precipitation Climatology')
#plt.savefig('/homes/areed29/CLIM997/summer22_research/precip_climo')
```

```python
blizzard_2016=daily_values.loc['2016-01-23']
blizzard_2016
```

```python
fig = plt.figure(figsize=(11,8.5))
ax=plt.axes(projection=ccrs.Robinson())
cs=ax.contourf(lon, lat, blizzard_2016,transform = ccrs.PlateCarree(),cmap='YlGnBu',extend='both')
ax.set_extent([-124.5,-66.8,23.486,49])
ax.coastlines()
ax.gridlines()
cbar=plt.colorbar(cs, shrink=0.75)
plt.title('Blizzard 2016 Precipitation (Janurary 23rd, 2016)')
```
