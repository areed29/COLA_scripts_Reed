```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
#from intake import open_catalog
import cartopy.io.shapereader as shpreader
#import mygrads as mg
#import Ngl
#import Nio
#import fiona
import matplotlib as mpl

```

```python
DS=xr.open_dataset('/project/xprecip/merra2_ncfiles/precip/resampled/all_merra_precip_6hourly_7-12.nc')
print(DS)
precip=DS.PRECTOTCORR
print(precip)
ts_1_prcp=DS.time
data_lat_precip=DS.lat.values
data_lon_precip=DS.lon.values
nlats_precip=len(data_lat_precip)
nlons_precip=len(data_lon_precip)
```

```python
winter_selected=precip.isel(time=precip.time.dt.season.isin(['DJF']))
winter_selected
```

```python
hourly_groups=winter_selected.groupby('time.hour').mean(dim='time')
hourly_groups
```

```python
fig = plt.figure(figsize=(11,8.5))
ax=plt.axes(projection=ccrs.Robinson())
cs=ax.contourf(data_lon_precip, data_lat_precip, hourly_groups[0,:,:],transform = ccrs.PlateCarree(),cmap='YlGnBu',extend='both')
ax.set_extent([-124.5,-66.8,23.486,49])
ax.coastlines()
ax.gridlines()
cbar=plt.colorbar(cs, shrink=0.75)
plt.title('DJF 00z Precipitation Climatology')
plt.savefig('/homes/areed29/CLIM997/summer22_research/DJF_seasonal_precip_climo_00z')
```

```python
fig = plt.figure(figsize=(11,8.5))
ax=plt.axes(projection=ccrs.Robinson())
cs=ax.contourf(data_lon_precip, data_lat_precip, hourly_groups[1,:,:],transform = ccrs.PlateCarree(),cmap='YlGnBu',extend='both')
ax.set_extent([-124.5,-66.8,23.486,49])
ax.coastlines()
ax.gridlines()
cbar=plt.colorbar(cs, shrink=0.75)
plt.title('DJF 06z Precipitation Climatology')
plt.savefig('/homes/areed29/CLIM997/summer22_research/DJF_seasonal_precip_climo_06z')
```

### Adding more samples to each hour

```python
hour0=winter_selected.isel(time=winter_selected.time.dt.hour.isin([0]))
hour0
```

```python
hour0_grouped=hour0.groupby('time.hour')
hour0_grouped
```

```python
print(hour0.time)
```

```python
window=3
rolling_mean0=hour0.rolling(time=window).mean(dim='time')
rolling_mean0
```

```python
climo_00z=rolling_mean0.groupby('time.hour').mean(dim='time')
climo_00z
```

```python
hour6=winter_selected.isel(time=winter_selected.time.dt.hour.isin([6]))
hour6
```

```python
window=3
rolling_mean6=hour6.rolling(time=window).mean(dim='time')
rolling_mean6
```

```python
climo_06z=rolling_mean6.groupby('time.hour').mean(dim='time')
climo_06z
```

```python
hour12=winter_selected.isel(time=winter_selected.time.dt.hour.isin([12]))
hour12
```

```python
window=3
rolling_mean12=hour12.rolling(time=window).mean(dim='time')
rolling_mean12
```

```python
climo_12z=rolling_mean12.groupby('time.hour').mean(dim='time')
climo_12z
```

```python
hour18=winter_selected.isel(time=winter_selected.time.dt.hour.isin([18]))
hour18
```

```python
window=3
rolling_mean18=hour18.rolling(time=window).mean(dim='time')
rolling_mean18
```

```python
climo_18z=rolling_mean18.groupby('time.hour').mean(dim='time')
climo_18z
```

```python
climo_0=rolling_mean0.isel(time=rolling_mean0.time.dt.year.isin(['1980']))
climo_0
```

```python
climo_concat=xr.concat([climo_00z,climo_06z,climo_12z,climo_18z],dim='hour')
climo_concat
```

```python
dc_00z_climo=climo_concat.sel(lat=39,lon=-77,method='nearest')
plt.plot(dc_00z_climo['hour'],dc_00z_climo)
plt.title('Washington DC DJF Precipitation Climatology at 00z over time')
plt.legend(['Seasonal DJF 00z Climatology'])
plt.savefig('/homes/areed29/CLIM997/summer22_research/washington_precip_seasonalDJFclimo_00z_3daymean')
```

### Anomaly Testing

```python
anoms_00z=hour0-climo_00z
anoms_00z
```

```python
# Normalize by standard deviation
norm_anoms_00z=anoms_00z/(anoms_00z.std())
norm_anoms_00z
```

```python
norm_anoms_00z.to_netcdf('/project/xprecip/merra2_ncfiles/precip/anomalies/norm_anoms_precip_00z_corrected.nc')
```

```python
dc_00z_anoms=norm_anoms_00z.sel(lat=39,lon=-77,method='nearest')
plt.plot(dc_00z_anoms['time'],dc_00z_anoms)
plt.axhline(y=0.0, color='r', linestyle='-')
plt.title('Washington DC Precipitation Seasonal DJF Normalized Anomalies for 00z')
plt.legend(['Seasonal DJF 00z Normalized Anomalies'])
plt.savefig('/homes/areed29/CLIM997/summer22_research/washington_precip_seasonalDJFnormanoms_00z')
```
