```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
#from intake import open_catalog
import cartopy.io.shapereader as shpreader
#import mygrads as mg
#import Ngl
#import Nio
#import fiona
import matplotlib as mpl
```

```python
DS=xr.open_dataset('/project/xprecip/merra2_ncfiles/winds/vort_computed_corrected_8-16.nc')
DS
vort=DS['__xarray_dataarray_variable__']
vort

```

```python
data_lat_vort=vort['lat']
data_lon_vort=vort['lon']
nlats_vort=len(data_lat_vort)
nlons_vort=len(data_lon_vort)
print(nlats_vort)
print(nlons_vort)
vort_time=vort['time']
vort_time
```

```python
winter_selected_vort=vort.isel(time=vort.time.dt.season.isin(['DJF']))
winter_selected_vort
```

```python
hourly_groups_vort=winter_selected_vort.groupby('time.hour').mean(dim='time')
hourly_groups_vort
```

```python
fig = plt.figure(figsize=(11,8.5))
ax=plt.axes(projection=ccrs.Robinson())
cs=ax.contourf(data_lon_vort, data_lat_vort, hourly_groups_vort[0,:,:]*100000,transform = ccrs.PlateCarree(),cmap='bwr',extend='both')
ax.set_extent([-124.5,-66.8,23.486,49])
ax.coastlines()
ax.gridlines()
cbar=plt.colorbar(cs, shrink=0.75)
plt.title('DJF 00z Vorticity Climatology')
plt.savefig('/homes/areed29/CLIM997/summer22_research/DJF_seasonal_vort_climo_00z')
```

### More robust climatology for vorticity utilizing boxcar filter

```python
hour0_vort=winter_selected_vort.isel(time=winter_selected_vort.time.dt.hour.isin([0]))
hour0_vort
```

```python
window=3
rolling_mean0_vort=hour0_vort.rolling(time=window).mean(dim='time')
rolling_mean0_vort
```

```python
climo_00z_vort=rolling_mean0_vort.groupby('time.hour').mean(dim='time')
climo_00z_vort
```

```python
hour6_vort=winter_selected_vort.isel(time=winter_selected_vort.time.dt.hour.isin([6]))
hour6_vort
```

```python
window=3
rolling_mean6_vort=hour6_vort.rolling(time=window).mean(dim='time')
rolling_mean6_vort
```

```python
climo_06z_vort=rolling_mean6_vort.groupby('time.hour').mean(dim='time')
climo_06z_vort
```

```python
hour12_vort=winter_selected_vort.isel(time=winter_selected_vort.time.dt.hour.isin([12]))
hour12_vort
```

```python
window=3
rolling_mean12_vort=hour12_vort.rolling(time=window).mean(dim='time')
rolling_mean12_vort
```

```python
climo_12z_vort=rolling_mean12_vort.groupby('time.hour').mean(dim='time')
climo_12z_vort
```

```python
hour18_vort=winter_selected_vort.isel(time=winter_selected_vort.time.dt.hour.isin([18]))
hour18_vort
```

```python
window=3
rolling_mean18_vort=hour18_vort.rolling(time=window).mean(dim='time')
rolling_mean18_vort
```

```python
climo_18z_vort=rolling_mean18_vort.groupby('time.hour').mean(dim='time')
climo_18z_vort
```

```python
vort_climo_concat=xr.concat([climo_00z_vort,climo_06z_vort,climo_12z_vort,climo_18z_vort],dim='hour')
vort_climo_concat
```

```python
dc_hourly_climo_vort=vort_climo_concat.sel(lat=39,lon=-77,method='nearest')
plt.plot(dc_hourly_climo_vort['hour'],dc_hourly_climo_vort*100000)
plt.axhline(y=0.0, color='r', linestyle='-')
plt.title('Washington DC DJF hourly Vorticity Climatology')
plt.legend(['Seasonal DJF Climatology'])
plt.savefig('/homes/areed29/CLIM997/summer22_research/washington_vort_seasonalDJFclimo_3daymean')
```

```python
dc_06z_climo_vort=rolling_mean6_vort.sel(lat=39,lon=-77,method='nearest')
plt.plot(dc_06z_climo_vort['time'],dc_06z_climo_vort*100000)
plt.title('Washington DC DJF Vorticity Climatology at 06z over time')
plt.axhline(y=0.0, color='r', linestyle='-')
plt.legend(['Seasonal DJF 06z Climatology'])
plt.savefig('/homes/areed29/CLIM997/summer22_research/washington_vort_seasonalDJFclimo_06z_3daymean')
```

```python
dc_12z_climo_vort=rolling_mean12_vort.sel(lat=39,lon=-77,method='nearest')
plt.plot(dc_12z_climo_vort['time'],dc_12z_climo_vort*100000)
plt.axhline(y=0.0, color='r', linestyle='-')
plt.title('Washington DC DJF Vorticity Climatology at 12z over time')
plt.legend(['Seasonal DJF 12z Climatology'])
plt.savefig('/homes/areed29/CLIM997/summer22_research/washington_vort_seasonalDJFclimo_12z_3daymean')
```

```python
dc_18z_climo_vort=rolling_mean18_vort.sel(lat=39,lon=-77,method='nearest')
plt.plot(dc_18z_climo_vort['time'],dc_18z_climo_vort*100000)
plt.axhline(y=0.0, color='r', linestyle='-')

plt.title('Washington DC DJF Vorticity Climatology at 18z over time')
plt.legend(['Seasonal DJF 18z Climatology'])
plt.savefig('/homes/areed29/CLIM997/summer22_research/washington_vort_seasonalDJFclimo_18z_3daymean')
```

### Normalized Anomalies

```python
anoms_00z_vort=hour0_vort-climo_00z_vort
anoms_00z_vort
```

```python
# Normalize by standard deviation
norm_anoms_00z_vort=anoms_00z_vort/(anoms_00z_vort.std())
norm_anoms_00z_vort
```

```python
norm_anoms_00z_vort.to_netcdf('/project/xprecip/merra2_ncfiles/winds/vort_analysis/norm_anoms_vort_00z_corrected.nc')
```

```python
dc_00z_anoms_vort=norm_anoms_00z_vort.sel(lat=39,lon=-77,method='nearest')
plt.plot(dc_00z_anoms_vort['time'],dc_00z_anoms_vort)
plt.axhline(y=0.0, color='r', linestyle='-')
plt.title('Washington DC Vorticity Seasonal DJF Normalized Anomalies for 00z')
plt.legend(['Seasonal DJF 00z Normalized Anomalies'])
plt.savefig('/homes/areed29/CLIM997/summer22_research/washington_vort_seasonalDJFnormanoms_00z')
```
