```python
#wind data for 300mb
f='/project/xprecip/merra2_ncfiles/winds/vwinds_preslevs_4-23/300/vwinds_300_winter.nc'
DS=xr.open_dataset(f)
vwinds=DS['V']
vwinds
```

```python
#Temperature
f='/project/xprecip/merra2_ncfiles/temp_pressurelevels/850/*.nc'
DS=xr.open_mfdataset(f,combine='nested',concat_dim='time')
DS
```

```python
winter_only_temp=DS['T'].isel(time=DS.time.dt.season.isin(['DJF'])).sel(lev=300)
winter_only_temp
```

```python
conustemp=winter_only_temp.sel(lat=slice(24.5,51.5),lon=slice(-126.25,-56.25))
conustemp
```
