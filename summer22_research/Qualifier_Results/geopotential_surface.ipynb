```python
f='/project/xprecip/merra2_ncfiles/geo_height/surface/resampled/high_pass/hpfilter_geoheight.nc'
DS_trans=xr.open_dataset(f)
#DS_trans
trans_geo=DS_trans['__xarray_dataarray_variable__']
trans_geo
```

```python
#wind by transients at 300 and 850
```

```python
f='/project/xprecip/merra2_ncfiles/winds/vwinds_preslevs_4-23/850/high_pass/850winds_highpass_0.75deg.nc'
DS=xr.open_dataset(f)
hp_850winds=DS['__xarray_dataarray_variable__']
hp_850winds
```

```python
f='/project/xprecip/merra2_ncfiles/winds/vwinds_preslevs_4-23/300/high_pass/hpfilter_vwind_0.75.nc'
DS=xr.open_dataset(f)
hp_300winds=DS['__xarray_dataarray_variable__']
hp_300winds
```

```python
geo_flux_trans_300=(trans_geo*hp_300winds).mean(dim='time')
geo_flux_trans_850=(trans_geo*hp_850winds).mean(dim='time')
```

```python
print(np.shape(geo_flux_trans_850))
```

```python
geo_early=trans_geo.sel(time=slice('1980-12-01','2001-02-28'))
geo_late=trans_geo.sel(time=slice('2001-12-01','2022-02-28'))
winds_early300=hp_300winds.sel(time=slice('1980-12-01','2001-02-28'))
winds_late300=hp_300winds.sel(time=slice('2001-12-01','2022-02-28'))
winds_early850=hp_850winds.sel(time=slice('1980-12-01','2001-02-28'))
winds_late850=hp_850winds.sel(time=slice('2001-12-01','2022-02-28'))
```

```python
geo_flux_trans_300early=(geo_early*winds_early300).mean(dim='time')
geo_flux_trans_850early=(geo_early*winds_early850).mean(dim='time')
geo_flux_trans_300late=(geo_late*winds_late300).mean(dim='time')
geo_flux_trans_850late=(geo_late*winds_late850).mean(dim='time')
```

```python
geo_flux_trans_300.to_netcdf('/project/xprecip/merra2_ncfiles/energy_calcs/corrections/geo_trans300.nc')
geo_flux_trans_850.to_netcdf('/project/xprecip/merra2_ncfiles/energy_calcs/corrections/geo_trans850.nc')
geo_flux_trans_300early.to_netcdf('/project/xprecip/merra2_ncfiles/energy_calcs/corrections/geo_trans300_early.nc')
geo_flux_trans_850early.to_netcdf('/project/xprecip/merra2_ncfiles/energy_calcs/corrections/geo_trans850_early.nc')
geo_flux_trans_300late.to_netcdf('/project/xprecip/merra2_ncfiles/energy_calcs/corrections/geo_trans300_late.nc')
geo_flux_trans_850late.to_netcdf('/project/xprecip/merra2_ncfiles/energy_calcs/corrections/geo_trans850_late.nc')
```
