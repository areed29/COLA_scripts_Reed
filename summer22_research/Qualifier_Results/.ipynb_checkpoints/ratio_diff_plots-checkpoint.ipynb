### TOTAL

```python
f='/project/xprecip/merra2_ncfiles/energy_calcs/sh_diff_total.nc'
DS_sh_diff=xr.open_dataset(f)
sh_diff=DS_sh_diff['__xarray_dataarray_variable__']
sh_diff
```

```python
f='/project/xprecip/merra2_ncfiles/energy_calcs/lh_diff_total.nc'
DS_lh_diff=xr.open_dataset(f)
lh_diff=DS_lh_diff['__xarray_dataarray_variable__']
lh_diff
```

### TRANSIENTS

```python
f='/project/xprecip/merra2_ncfiles/energy_calcs/sh_diff_transients850.nc'
DS_sh_diff=xr.open_dataset(f)
sh_diff_transients=DS_sh_diff['__xarray_dataarray_variable__']
sh_diff_transients
```

```python
f='/project/xprecip/merra2_ncfiles/energy_calcs/lh_diff_transients850.nc'
DS_lh_diff=xr.open_dataset(f)
lh_diff_transients=DS_lh_diff['__xarray_dataarray_variable__']
lh_diff_transients
```

```python
f='/project/xprecip/merra2_ncfiles/energy_calcs/geofluxtransients_diff850.nc'
DS_geoflux_diff=xr.open_dataset(f)
geoflux_diff_transients=DS_geoflux_diff['__xarray_dataarray_variable__']
geoflux_diff_transients
```

```python
### RATIO OF TRANSIENT DIFF/TOTAL DIFF
```

```python
diff_fraction_sh=(sh_diff_transients/sh_diff)*100
diff_fraction_sh
```

```python
diff_fraction_lh=(lh_diff_transients/lh_diff)*100
diff_fraction_lh
```

```python
geoflux_fraction=(geoflux_diff_transients/geoflux_diff)*100
geoflux_fraction
```

```python
diff_all=[diff_fraction_sh,diff_fraction_lh,geoflux_fraction]
diff_all
```

```python
# Define the figure and each axis for the 2 rows and 2 columns
fig, axs = plt.subplots(nrows=3,ncols=1,
                        subplot_kw={'projection': ccrs.PlateCarree()},
                        figsize=(20,10))
titles=['Sensible Heat Transport', 'Latent Heat transport','Geopotential flux transport']
# axs is a 2 dimensional array of `GeoAxes`.  
# We will flatten it into a 1-D array
axs=axs.flatten()
us_shapes = list(shpreader.Reader("/scratch/areed29/gadm36_USA_shp/gadm36_USA_1.shp").geometries())

#Loop over all of the seasons and plot
#for i in enumerate(climo_all):
for i in range(len(diff_all)):
        # Select the season
        data=diff_all[i]
        # Add the cyclic point
        data,lons=add_cyclic_point(data,coord=diff_fraction_sh['lon'])
        max_val=100
        increment=20
        clevs = np.arange(start=-max_val, stop=max_val+increment, step=increment)
        # Contour plot
        #norm=colors.SymLogNorm(linthresh=0.03, linscale=0.03,
        #                                      vmin=-1600000, vmax=1600000, base=10)
        cs=axs[i].contourf(lons,diff_fraction_sh['lat'],data,clevs,
                          transform = ccrs.PlateCarree(),
                          cmap='BrBG',extend='both')
       # Longitude labels
        axs[i].set_xticks(np.arange(-180,181,60), crs=ccrs.PlateCarree())
        lon_formatter = cticker.LongitudeFormatter()
        axs[i].xaxis.set_major_formatter(lon_formatter)

        # Latitude labels
        axs[i].set_yticks(np.arange(-90,91,30), crs=ccrs.PlateCarree())
        lat_formatter = cticker.LatitudeFormatter()
        axs[i].yaxis.set_major_formatter(lat_formatter)

        # Title each subplot with the name of the season
        axs[i].set_title(titles[i])

        # Draw the coastines for each subplot
        axs[i].coastlines()
        axs[i].set_extent([-124.5,-66.8,23.486,49])
        axs[i].add_geometries(us_shapes, ccrs.PlateCarree(), edgecolor='gray',
                  facecolor='none')
# Adjust the location of the subplots 
# on the page to make room for the colorbar
fig.subplots_adjust(bottom=0.25, top=0.9, left=0.05, right=0.95,
                    wspace=0.2, hspace=0.2)

# Add a colorbar axis at the bottom of the graph
cbar_ax = fig.add_axes([0.2, 0.175, 0.6, 0.02])

# Draw the colorbar
cbar=fig.colorbar(cs, cax=cbar_ax,orientation='horizontal',label='percentage of difference (m^3/s^3)')
fig.tight_layout()
# Add a big title at the top
plt.suptitle('Percentage of Energy Transport Late-Early Climatology Difference from transients (850mb)',fontsize=15)
plt.savefig('/homes/areed29/CLIM997/summer22_research/Qualifier_Results/850mb_percentagedifftransients.png')
```
