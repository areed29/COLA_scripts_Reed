```python
import numpy as np
import scipy.signal as sig

def high_pass_filter(data, cutoff_freq, sampling_rate):
    nyquist_freq = sampling_rate / 2
    norm_cutoff_freq = cutoff_freq / nyquist_freq
    b, a = sig.butter(2, norm_cutoff_freq, btype='highpass')
    filtered_data = sig.filtfilt(b, a, data)
    return filtered_data
```

```python
#temp_restricted=temp850.sel(time=slice('1980-12-01','1984-02-28'))
hp_restricted=temp850.sel(lat=slice(24.5,51.5),lon=slice(-126.25,-56.25))
#hp_restricted
hpall=[]
for k in range(len(hp_restricted['lat'])):
    for j in range(len(hp_restricted['lon'])):
        hpall.append(high_pass_filter(hp_restricted[:,k,j],0.0000011574,10))
hpall
```

```python
print(np.shape(hpall)) #space,time
```

```python
len_time=len(hp_restricted['time'])
#len_time
len_lat=len(hp_restricted['lat'])
len_lon=len(hp_restricted['lon'])
```

```python
hp_reshape=np.reshape(hpall,(len_time,len_lat,len_lon))
#hp_reshape
print(np.shape(hp_reshape))
```

```python
hp_array=xr.DataArray(hp_reshape, coords={'time':time,
                                           'lat':hp_restricted['lat'].values, 'lon':hp_restricted['lon'].values},
                        dims=['time','lat','lon'])
hp_array
```
