### Time mean of u and v wind at 850mb

```python
f='/project/xprecip/merra2_ncfiles/winds/MERRA2_400.inst6_3d_ana_Np.20211026.SUB.nc'
#DS= xr.open_mfdataset(f,combine='nested',concat_dim='time')
DS= xr.open_dataset(f)
print(DS)
ts_1=DS.time
```

```python
lon=DS.lon.values
lat=DS.lat.values
print(np.shape(lon))
```

```python
u_wind= DS.U
v_wind = DS.V
u_wind_mean=u_wind.sel(lev='850.0').mean(dim=('time'))
v_wind_mean=v_wind.sel(lev='850.0').mean(dim=('time'))
u_wind_mean
```

### 850mb Vorticity using myGRADS

```python
vort = mg.hcurl(u_wind_mean,v_wind_mean,lat,lon)
vort
new_vort=vort*100000
```

```python

fig = plt.figure(figsize=(11,8.5))
ax=plt.axes(projection=ccrs.Robinson())

cs=ax.contourf(lon, lat, new_vort,transform = ccrs.PlateCarree(),cmap='RdBu_r',extend='both')
ax.set_extent([-124.5,-66.8,23.486,49])
ax.coastlines()
ax.gridlines()
cbar=plt.colorbar(cs, shrink=0.75,label='[$x10^{-5}$ s$^{-1}$]')
#superimposing wind vectors
ax.quiver(lon, lat, u_wind_mean.values, v_wind_mean.values, minlength=0.1, scale_units='xy',scale=0.0001, transform=ccrs.PlateCarree(), color='k',width=0.001)
plt.title('Relative Vorticity with wind vectors at 850mb')
plt.savefig('/homes/areed29/CLIM997/summer22_research/conus_vort_winds_10-26-21')
```
