### Prototyping- create 6hourly sums using 1 data file first

```python
f='/project/xprecip/merra2_ncfiles/precip/MERRA2_401.tavg1_2d_flx_Nx.20210825.SUB.nc'
DS= xr.open_dataset(f)
print(DS)
precip=DS.PRECTOTCORR
```

```python
precip['time']
```

```python
#Point to check: Washington, DC: lat/lon 39, -77
hourly_pt=precip.sel(lat=39,lon=-77,method='nearest')
hourly_pt_vals=hourly_pt.values
print(hourly_pt_vals)
```

```python
#Calculating manually:
six_hourly=np.sum(hourly_pt_vals[0:6])
six_hourly #correct result when checked manually!
```

```python
six_hourly_data=hourly_pt.resample(time="6H").sum(dim='time')
six_hourly_data.values
#first result is, in fact, the first 6 values summed!
```

### Does Washington DC recieve 1m/1000mm per year on average?

```python
yearly_average=DS1.groupby('time.year').mean(dim='time')
yearly_average
```

```python
precip_avg=yearly_average['PRECTOTCORR']
mean_precip=np.nanmean(precip_avg,axis=0)
mean_precip
```

```python
fig = plt.figure(figsize=(11,8.5))
ax=plt.axes(projection=ccrs.Robinson())
cs=ax.contourf(yearly_average['lon'], yearly_average['lat'], mean_precip,transform = ccrs.PlateCarree(),cmap='YlGnBu',extend='both')
ax.set_extent([-124.5,-66.8,23.486,49])
ax.coastlines()
ax.gridlines()
cbar=plt.colorbar(cs, shrink=0.75)
plt.title('Precipitation Climatology')
#plt.savefig('/homes/areed29/CLIM997/summer22_research/precip_climo')
```

```python
yearly_sum=mean_precip*365*86400
yearly_sum
```

```python
fig = plt.figure(figsize=(11,8.5))
ax=plt.axes(projection=ccrs.Robinson())
cs=ax.contourf(yearly_average['lon'], yearly_average['lat'], yearly_sum,transform = ccrs.PlateCarree(),cmap='YlGnBu',extend='both')
ax.set_extent([-124.5,-66.8,23.486,49])
ax.coastlines()
ax.gridlines()
cbar=plt.colorbar(cs, shrink=0.75)
plt.title('Precipitation Climatology')
```
