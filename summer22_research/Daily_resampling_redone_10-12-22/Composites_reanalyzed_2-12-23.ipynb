```python
nlats_precip=len(lat_precip)
nlons_precip=len(lon_precip)
rad = 4.*np.arctan(1.)/180.
clat = np.cos(lat_precip*rad)
print(clat)
```

```python
precip_bounded=[]
bounds=np.full_like((dec_precip_anoms),np.nan)
for k in range(nlats_precip):
    for j in range(nlons_precip):
        if (lat_precip_np[k]<=max_lat and lat_precip_np[k]>=min_lat and lon_precip_np[j]<=max_lon and lon_precip_np[j]>=min_lon):
            bounds[:,k,j]=1
            precip_bounded_dec=dec_precip_anoms*bounds
```
