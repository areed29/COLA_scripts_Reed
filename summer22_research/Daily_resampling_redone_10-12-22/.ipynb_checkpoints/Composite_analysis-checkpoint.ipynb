### First experiment: Bounding box the size of Virginia


```python
max_lat=39.5
min_lat=36.5
max_lon=-75.625
min_lon=-83.75
```

```python
nlats_vort=len(lat_vort)
nlons_vort=len(lon_vort)
rad = 4.*np.arctan(1.)/180.
clat = np.cos(lat_vort*rad)
print(clat)
```

```python
lat_vort_np=lat_vort.values
lon_vort_np=lon_vort.values
```

```python
vort_bounded=[]
bounds=np.full_like((vort_anoms),np.nan)
for k in range(nlats_vort):
    for j in range(nlons_vort):
        if (lat_vort_np[k]<=max_lat and lat_vort_np[k]>=min_lat and lon_vort_np[j]<=max_lon and lon_vort_np[j]>=min_lon):
            bounds[:,k,j]=1
            vort_bounded=vort_anoms*bounds
```

```python
fig = plt.figure(figsize=(11,8.5))
ax=plt.axes(projection=ccrs.Robinson())
cs=ax.contourf(vort_bounded['lon'], vort_bounded['lat'], vort_bounded[0,:,:] ,transform = ccrs.PlateCarree(),cmap='RdBu_r',extend='both')
ax.set_extent([-124.5,-66.8,23.486,49])
ax.coastlines()
# Add gridlines
ax.gridlines()
us_shapes = list(shpreader.Reader("/scratch/areed29/gadm36_USA_shp/gadm36_USA_1.shp").geometries())
ax.add_geometries(us_shapes, ccrs.PlateCarree(), edgecolor='gray',
                  facecolor='none')

cbar=plt.colorbar(cs, shrink=0.75)
plt.title('Bounding box visualization')
```

```python
#Without weighting
non_weighted_mean=vort_bounded.mean(dim=('lat','lon'))
non_weighted_mean
```

```python
plt.plot(non_weighted_mean['time'],non_weighted_mean/non_weighted_mean.std(dim='time'))
plt.title('Normalized non-weighted mean Vorticity anomalies for VA bounding box')
```

```python
norm=non_weighted_mean/non_weighted_mean.std(dim='time')
norm.std(dim='time')
```

```python
#Applying the weights
weighted_mean=vort_bounded.weighted(weights=clat).mean(dim=('lat','lon'))
```

```python
plt.plot(weighted_mean['time'],weighted_mean)
plt.title('Weighted mean of DJF Vorticity Anomalies for VA bounding box')
```

```python
normalized=weighted_mean/(weighted_mean.std(dim='time'))
normalized
```

```python
print(normalized.mean(dim='time'))
print(np.std(normalized,axis=0))
```

```python
plt.plot(normalized['time'],normalized)
plt.title('Weighted Mean of DJF Normalized Vorticity Anomalies for VA bounding box')
```

```python
#Define an ETC as any normalized vorticity anomaly greater than 1 (1 standard deviation above mean)
etc_va=normalized.where(normalized>1.)
etc_va
```

```python
plt.plot(etc_va['time'],etc_va)
plt.title('DJF Normalized Vorticity Anomalies for VA bounding box greater than 1 SD above mean (ETC)')
```

```python
etc_va_sametime=etc_va.sel(time=slice('1980-01-01','2021-02-28'))
etc_va_sametime
```

```python
va_vort_precip=norm_precip_anoms.sel(time=etc_va_sametime.dropna(dim='time')['time']).mean(dim='time')
va_vort_precip
```

```python
fig = plt.figure(figsize=(11,8.5))
ax=plt.axes(projection=ccrs.Robinson())
#norm = mpl.colors.Normalize(vmin=-1.,vmax=1.)
levels=np.linspace(-1,1)
cs=ax.contourf(va_vort_precip['lon'], va_vort_precip['lat'], va_vort_precip ,levels,transform = ccrs.PlateCarree(),cmap='BrBG',extend='both')
ax.set_extent([-124.5,-66.8,23.486,49])
ax.coastlines()
# Add gridlines
ax.gridlines()
us_shapes = list(shpreader.Reader("/scratch/areed29/gadm36_USA_shp/gadm36_USA_1.shp").geometries())
ax.add_geometries(us_shapes, ccrs.PlateCarree(), edgecolor='gray',
                  facecolor='none')

cbar=plt.colorbar(cs, shrink=0.75)
plt.title('Composite Normalized Precipitation Anomalies for VA ETC vorticities')
#plt.savefig('/homes/areed29/CLIM997/summer22_research/composite_anoms_vavorts')
```

```python
rad = 4.*np.arctan(1.)/180.
clat = np.cos(lat_precip*rad)
print(clat)
```

```python
precip_bounded=[]
bounds=np.full_like((precip_anoms),np.nan)
for k in range(nlats_precip):
    for j in range(nlons_precip):
        if (lat_precip_np[k]<=max_lat and lat_precip_np[k]>=min_lat and lon_precip_np[j]<=max_lon and lon_precip_np[j]>=min_lon):
            bounds[:,k,j]=1
            precip_bounded=precip_anoms*bounds
```
